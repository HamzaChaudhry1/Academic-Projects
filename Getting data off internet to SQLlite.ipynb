```python
# _*_ coding utf-8 _*_

"""

entityid
entityname 
entitystatus
principle zip code
entity type

@ Author: Muhammad Chaudhry
"""
import sqlite3
import json
import urllib.request, urllib.parse, urllib.error

conn = sqlite3.connect("cobiz.sqlite3")
cursor = conn.cursor()

qry = "Drop Table CoBizData"
cursor.execute(qry)


qry = """CREATE TABLE IF NOT EXISTS COBizData (entityid INTEGER Primary Key,
entityname TEXT, entitystatus TEXT, principalzipcode TEXT,
entitytype TEXT)"""

cursor.execute(qry)

url = "https://data.colorado.gov/resource/4ykn-tg5h.json?principalzipcode=80202&$limit=10000"
document = urllib.request.urlopen(url)

#get all of the text from the document
text = document.read().decode()

jsdata = json.loads(text)

count = 0

for data in jsdata:
    qry = """INSERT INTO COBizData Values(?, ?, ?, ?, ?)"""

    cursor.execute(qry, (data["entityid"], data["entityname"], data["entitystatus"], data["principalzipcode"], data["entitytype"]))
    count = count + 1
    if count % 100 == 0:
        conn.commit()
        print(count)
    
    
conn.commit()

qry = "SELECT * FROM CoBizData WHERE entitystatus <> 'Good Standing'"
cursor.execute(qry)

for row in cursor:
    print(row)

conn.close()
```
